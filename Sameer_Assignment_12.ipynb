```python
import pandas as pd
```

```python
data = pd.read_csv('retail supermarket.csv')
```

## ***Question No.1***

### 1- Find record if Sales greater than 500 and Quantity greater than 5?

## ***Answer No.1***

```python
data[(data['Sales'] > 500) & (data['Quantity'] > 5)]
```

####  There are 431 records with sales exceeding 500 and quantity exceeding 5.

## ***Question No.2***

### 2- Find records if Discount is more than 0.5 (50%) or Profit is negative.

## ***Answer No.2***

```python
data[(data['Discount'] > 0.5) | (data['Profit'] < 1)]
```

####  There are 2202 records with a discount greater than 50% or negative profit.

## ***Question No.3***

### 3- Display Ship Mode and Profit columns if region is East using loc

## ***Answer No.3***

```python
data.loc[data['Region'] == 'East', ['Ship Mode','Profit']]
```

####  There are 2848 records showing Ship Mode and Profit for the East region.

## ***Question No.4***

### 4- How many records belong to the "Corporate" segment and the "Central" region?

## ***Answer No.4***

```python
data[(data['Segment'] == 'Corporate') & (data['Region'] == 'Central')]
```

####  There are 673 records for Corporate segment in the Central region.

## ***Question No.5***

### 5- Find all orders in the "Furniture" category but not in "Chairs" sub-category.

## ***Answer No.5***

```python
data[(data['Category'] == 'Furniture') & ~(data['Sub-Category'] == 'Chairs')]
```

####  There are 1504 orders in the Furniture category that are not in the Chairs sub-category.

## ***Question No.6***

### 6- Which orders have discount (>0.3) and still earned profit (>0)?

## ***Answer No.6***

```python
data[(data['Discount'] > 0.3) & (data['Profit'] > 0)]
```

####  There are 26 orders that have a discount greater than 30% and still earned a positive profit.

## ***Question No.7***

### 7- Find high-profit orders where Quantity > 5 and Category == 'Technology'.

## ***Answer No.7***

```python
data[(data['Quantity'] > 5) & (data['Category'] == 'Technology')]
```

####  There are 323 high-profit orders in the Technology category with quantities greater than 5.

## ***Question No.8***

### 8- Select rows 0â€“4 and columns 'Ship Mode', 'Segment', 'Sales' by position using iloc

## ***Answer No.8***

####  There are 26 orders that have a discount greater than 30% and still earned a positive profit.

```python
data.iloc[0:5, [0,1,9]]
```

## ***Question No.9***

### 9- Show Home Office segment orders with Sales > 1000 and Quantity > 3.

## ***Answer No.9***

```python
data[(data['Segment'] == 'Home Office') & (data['Sales'] > 1000) & (data['Quantity'] > 3)]
```

####  There are 73 Home Office segment orders with sales exceeding 1000 and quantities greater than 3.

## ***Question No.10***

### 10- Find orders where Profit Margin > 0.3

Profit Margin = (Profit/Sales)

## ***Answer No.10***

```python
data[(data['Profit'] / data['Sales']) > 0.3]
```

####  There are 4145 orders with a profit margin greater than 30%

```python
#Saving the data in 1 excel file with 10 Questions on different tabs.
```

```python
with pd.ExcelWriter('Sameer_Assignment_12.xlsx') as writer:
    
    # Question No.1
    data[(data['Sales'] > 500) & (data['Quantity'] > 5)].to_excel(writer, sheet_name='Question 1', index=False)
    
    # Question No.2
    data[(data['Discount'] > 0.5) | (data['Profit'] < 1)].to_excel(writer, sheet_name='Question 2', index=False)
    
    # Question No.3
    data.loc[data['Region'] == 'East', ['Ship Mode','Profit']].to_excel(writer, sheet_name='Question 3', index=False)
    
    # Question No.4
    data[(data['Segment'] == 'Corporate') & (data['Region'] == 'Central')].to_excel(writer, sheet_name='Question 4', index=False)
    
    # Question No.5
    data[(data['Category'] == 'Furniture') & ~(data['Sub-Category'] == 'Chairs')].to_excel(writer, sheet_name='Question 5', index=False)
    
    # Question No.6
    data[(data['Discount'] > 0.3) & (data['Profit'] > 0)].to_excel(writer, sheet_name='Question 6', index=False)
    
    # Question No.7
    data[(data['Quantity'] > 5) & (data['Category'] == 'Technology')].to_excel(writer, sheet_name='Question 7', index=False)
    
    # Question No.8
    data.iloc[0:5, [0,1,9]].to_excel(writer, sheet_name='Question 8', index=False)
    
    # Question No.9
    data[(data['Segment'] == 'Home Office') & (data['Sales'] > 1000) & (data['Quantity'] > 3)].to_excel(writer, sheet_name='Question 9', index=False)
    
    # Question No.10
    data[(data['Profit'] / data['Sales']) > 0.3].to_excel(writer, sheet_name='Question 10', index=False)
```
