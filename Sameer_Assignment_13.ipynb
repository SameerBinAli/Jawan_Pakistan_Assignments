```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Salaries.csv')
data
```

## ***1. Drop rows where EmployeeName and JobTitle are 'Not provided'***

```python
data = data[(data['EmployeeName'] != 'Not provided') & (data['JobTitle'] != 'Not provided')]
```

```python
data
```

## ***2. Replace 'Not Provided' with 0 in 'BasePay' 'OvertimePay'	'OtherPay'	'Benefits'***

```python
data[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']] = data[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']].replace('Not Provided', 0)
```

```python
data[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']]
```

## ***3. Change BasePay	OvertimePay	OtherPay	Benefits type to float***

```python
data[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']] = data[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']].astype(float)
```

```python
data[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']].dtypes
```

## ***4. Display all employees name and basepay columns who have a BasePay greater than 200,000.***

```python
data[data['BasePay'] > 200000][['EmployeeName', 'BasePay']]
```

## ***5. Find all employees whose JobTitle contains the word “Manager” (case = False)***

```python
data[data.JobTitle.str.contains('Manager', case=False)]
```

## ***6. Retrieve only the columns: EmployeeName, JobTitle, and TotalPay.***

```python
data[['EmployeeName', 'JobTitle','TotalPay']]
```

## ***7. Find count of employees who have zero OvertimePay.***

```python
data[data['OvertimePay'] == 0]
```

## ***8. Calculate the TotalPay for each JobTitle where totalpay > 300000***

```python
data[data['TotalPay'] > 300000].groupby('JobTitle')['TotalPay'].sum()
```

## ***9. Find the top 5 JobTitles with the highest average TotalPayBenefits show in descending order***

```python
data.groupby('JobTitle')['TotalPayBenefits'].mean().sort_values(ascending=False).head(5)
```

## ***10. Replace all missing Benefits values with 0***

```python
data['Benefits'] = data['Benefits'].replace(np.nan, 0)
```

```python
data['Benefits']
```

## ***11. Identify the top employee name by TotalPay in each Year. Show year, employeename and totalpay dataframe***

```python
data.loc[data.groupby('Year')['TotalPay'].idxmax()][['Year', 'EmployeeName', 'TotalPay']]
```
