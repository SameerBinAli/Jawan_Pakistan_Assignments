## ***Find upper bound, lower bound, number of outliers in Engine HP***

```python
Q1 = data['Engine HP'].quantile(0.25)
Q3 = data['Engine HP'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

outliers = data[(data['Engine HP'] < lower_bound) | (data['Engine HP'] > upper_bound)]['Engine HP']
```

```python
print(f"Upper bound: {upper_bound}\n")
print(f"Lower bound: {lower_bound}\n")
print(f"Number of outliers: {len(outliers)}")
```

## ***Fill missing values in Number of Doors column with median value then check if values are filled***

```python
median = data['Number of Doors'].median()
data['Number of Doors'].fillna(median, inplace=True)

print(f"Median value used for filling: {median}\n")
print(f"Missing values in 'Number Of Doors' after filling: {data['Number of Doors'].isnull().sum()}")
```
